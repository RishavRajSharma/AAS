```python
import random

# Simulated "training" (fusion rule from dataset concept)
# In real case, IMU + Vision data come from EuRoC MAV dataset
print("Fusion model initialized (mean fusion)")

# Option: use given values OR random values
choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    imu = 32
    vision = 28
else:
    imu = random.randint(0, 90)      # IMU angle
    vision = random.randint(0, 90)   # Vision angle

# Sensor fusion (mean)
fused = (imu + vision) / 2

# Output
print("IMU Angle:", imu, "°")
print("Vision Angle:", vision, "°")
print("Fused Angle:", fused, "°")
```
