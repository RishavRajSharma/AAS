```python
import random

desired_distance = 0.5
print("Wall Following Controller Initialized (Desired =", desired_distance, "m)")

choice = input("Enter '1' for given distance or '2' for random: ")

if choice == '1':
    actual_distance = 0.2
else:
    # Random distance (simulating sensor data)
    actual_distance = round(random.uniform(0.1, 1.0), 2)

# Control logic
if actual_distance < desired_distance:
    action = "Move Away from Wall"
else:
    action = "Maintain/Move Closer"

# Output
print("Desired Distance:", desired_distance, "m")
print("Actual Distance:", actual_distance, "m")
print("Action:", action)
```
