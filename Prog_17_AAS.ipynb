```python
import random
import math

# Simulated "training" (RRT goal connection rule)
goal_radius = 0.4
print("RRT Planner Initialized (Goal Radius =", goal_radius, ")")

# Option: use fixed or random node distance
choice = input("Enter '1' for near-goal case or '2' for random: ")

if choice == '1':
    distance_to_goal = 0.3   # within radius → connected
else:
    distance_to_goal = round(random.uniform(0.1, 1.0), 2)

# Check goal connection
if distance_to_goal <= goal_radius:
    result = "Goal Connected"
else:
    result = "Not Connected"

# Output
print("Distance to Goal:", distance_to_goal, "m")
print("Result:", result)
```
