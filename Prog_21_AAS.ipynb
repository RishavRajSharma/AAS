```python
import random

print("Reinforcement Learning Model Initialized")

# Option: use given values OR random values
choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    reward = 8
    learning_rate = 0.5
else:
    # Random values
    reward = random.randint(1, 10)
    learning_rate = round(random.uniform(0.1, 1.0), 2)

Q = reward * learning_rate

# Output
print("Reward:", reward)
print("Learning Rate:", learning_rate)
print("Q-value:", Q)
```
