```python
# Simulated data inspired by NASA Robot Motor Dataset
motor_data = {
    "distance_meters": 5.0,   # distance travelled in meters
    "time_seconds": 2.0       # time taken in seconds
}

# Step 1: Extract values
distance = motor_data["distance_meters"]
time = motor_data["time_seconds"]

# Step 2: Calculate speed
speed = distance / time

# Step 3: Display output
print("Distance:", distance, "meters")
print("Time:", time, "seconds")
print("Calculated Speed:", speed, "m/s")

# Final Output Requirement
if speed == 2.5:
    print("\nOutput: 2.5 m/s")
else:
    print("\nOutput:", speed, "m/s")
```
