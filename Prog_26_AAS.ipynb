```python
import random

print("Predictive Maintenance Model Initialized")

choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    T = 80   # Temperature
    V = 10   # Vibration
else:
    # Random values (simulating sensor data)
    T = random.randint(50, 100)
    V = random.randint(5, 20)

health = int(100 - (0.4 * T + 1.1 * V))

# Output
print("Temperature:", T)
print("Vibration:", V)
print("Health Score:", health)
```
