```python
import random

time_limit = 100
print("Logistics System Initialized (Time Limit =", time_limit, "s)")

choice = input("Enter '1' for given time or '2' for random: ")

if choice == '1':
    time_taken = 95
else:
    # Random delivery time
    time_taken = random.randint(50, 150)

# Delivery check
if time_taken <= time_limit:
    result = "Package Delivered"
else:
    result = "Delivery Delayed"

# Output
print("Time Taken:", time_taken, "s")
print("Result:", result)
```
