```python
import random

threshold = 0.3
print("Obstacle Avoidance System Initialized (Threshold =", threshold, "m)")

choice = input("Enter '1' for given distance or '2' for random: ")

if choice == '1':
    distance = 0.25
else:
    # Random distance (simulating sensor data)
    distance = round(random.uniform(0.1, 1.0), 2)

# Decision rule
if distance < threshold:
    action = "Turn Right"
else:
    action = "Move Forward"

# Output
print("Obstacle Distance:", distance, "m")
print("Action:", action)
```
