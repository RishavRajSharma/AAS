```python
import random

# Simulated "training" (path cost comparison)
print("Path Planning Model Initialized")

# Option: use given values OR random values
choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    path_A = 6
    path_B = 4
else:
    # Random costs (simulating dataset paths)
    path_A = random.randint(1, 10)
    path_B = random.randint(1, 10)

# Greedy selection (minimum cost)
if path_A < path_B:
    result = "Path A Selected"
else:
    result = "Path B Selected"

# Output
print("Path A Cost:", path_A)
print("Path B Cost:", path_B)
print("Result:", result)
```
