```python
import random

print("Visual SLAM pose update initialized")

choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    prev_pose = (2, 3)
    motion = (0.4, 0.6)
else:

    prev_pose = (random.randint(0, 5), random.randint(0, 5))
    motion = (round(random.uniform(0, 1), 2), round(random.uniform(0, 1), 2))

new_pose = (prev_pose[0] + motion[0], prev_pose[1] + motion[1])

# Output
print("Previous Pose:", prev_pose)
print("Motion:", motion)
print("New Pose:", new_pose)
```
