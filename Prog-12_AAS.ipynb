```python
import random

trained_threshold = 0.9

print("Model trained using dataset (threshold =", trained_threshold, ")")

choice = input("Enter '1' for given similarity or '2' for random: ")

if choice == '1':
    similarity = 0.94
else:
    similarity = round(random.uniform(0.5, 1.0), 2)  # random similarity score

if similarity > trained_threshold:
    result = "Place Recognized"
else:
    result = "Place Not Recognized"

# Output
print("Similarity Score:", similarity)
print("Result:", result)
```
