```python
import random

print("PID Controller Initialized")

# random values
choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    Kp = 0.12
    error = 12
else:
    # Random values
    Kp = round(random.uniform(0.05, 0.5), 2)
    error = random.randint(1, 20)

# Compute control output
u = round(Kp * error, 2)

# Output
print("Proportional Gain (Kp):", Kp)
print("Error:", error)
print("Control Output:", u)
```
