```python
import random

print("IMU Correction System Initialized")

choice = input("Enter '1' for given drift or '2' for random: ")

if choice == '1':
    drift = 1.8
else:
    # Random drift (in degrees)
    drift = round(random.uniform(0.5, 5.0), 2)

# Assume measured orientation
measured_orientation = 30.0  # degrees

# Apply correction
corrected_orientation = measured_orientation - drift

# Output
print("Measured Orientation:", measured_orientation, "°")
print("Drift:", drift, "°")
print("Corrected Orientation:", corrected_orientation, "°")
print("Result: Orientation Stabilized")
```
