```python
import random

print("Kalman Filter initialized (simple averaging)")

# Option: use given values OR random values
choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    prediction = 20
    measurement = 24
else:
    # Random values (simulating noisy sensor data)
    prediction = random.randint(10, 30)
    measurement = random.randint(10, 30)

# Kalman estimate (simplified)
estimate = (prediction + measurement) / 2

# Output
print("Prediction:", prediction)
print("Measurement:", measurement)
print("Estimated State:", estimate)
```
