```python
import time

class PublisherNode:
    def __init__(self):
        self.count = 0
        self.timer_period = 1.0 / 15.0  # 15 Hz

    def publisher_message(self):
        msg = f'Hello (Simulated ROS2): {self.count}'
        print(f'Publishing: {msg}')
        self.count += 1

    def run(self):
        try:
            while True:
                self.publisher_message()
                time.sleep(self.timer_period)
        except KeyboardInterrupt:
            print("\nNode stopped.")

def main():
    node = PublisherNode()
    node.run()

if __name__ == '__main__':
    main()
```
