```python
import random

# Simulated "training" (differential drive rule)
print("Kinematics Model Initialized")

# Option: use given values OR random values
choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    left = 3
    right = 3
else:
    # Random wheel speeds
    left = random.randint(0, 5)
    right = random.randint(0, 5)

# Determine motion type
if left == right:
    motion = "Straight Line Motion"
elif left > right:
    motion = "Turn Right"
else:
    motion = "Turn Left"

# Output
print("Left Wheel Speed:", left)
print("Right Wheel Speed:", right)
print("Motion:", motion)
```
