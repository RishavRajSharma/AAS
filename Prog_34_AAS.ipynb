```python
import random

print("Teleoperation System Initialized")

choice = input("Enter '1' to press a key or '2' for random key: ")

if choice == '1':
    key = input("Press a key (W/A/S/D): ").upper()
else:
    key = random.choice(['W', 'A', 'S', 'D'])

# Key mapping to actions
if key == 'A':
    action = "Left Turn"
elif key == 'D':
    action = "Right Turn"
elif key == 'W':
    action = "Move Forward"
elif key == 'S':
    action = "Move Backward"
else:
    action = "Invalid Key"

# Output
print("Key Pressed:", key)
print("Action:", action)
```
