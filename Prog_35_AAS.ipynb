```python
import random

threshold = 0.1
print("Localization System Initialized (Threshold =", threshold, "m)")

choice = input("Enter '1' for given error or '2' for random: ")

if choice == '1':
    pose_error = 0.08
else:
    # Random pose error (simulating localization output)
    pose_error = round(random.uniform(0.01, 0.2), 2)

# Accuracy check
if pose_error < threshold:
    result = "Accurate Localization"
else:
    result = "Localization Error High"

# Output
print("Pose Error:", pose_error, "m")
print("Result:", result)
```
