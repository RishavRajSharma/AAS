```python
import random

# Simulated "training" (rule-based logic)
print("Rule-Based Agent Initialized")

# Option: use fixed values OR random values
choice = input("Enter '1' for given condition or '2' for random: ")

if choice == '1':
    obstacle = True
    speed = 5
else:
    # Random conditions
    obstacle = random.choice([True, False])
    speed = random.randint(0, 10)

# Rule: If obstacle AND speed > 0 → Stop
if obstacle and speed > 0:
    action = "Stop"
else:
    action = "Move"

# Output
print("Obstacle Detected:", obstacle)
print("Speed:", speed)
print("Action:", action)
```
