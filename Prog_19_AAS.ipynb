```python
import random

print("Belief Update Model Initialized")

choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    prior = 0.7
    observation = 0.6
else:
    # Random probabilities (0 to 1)
    prior = round(random.uniform(0, 1), 2)
    observation = round(random.uniform(0, 1), 2)

# Bayesian belief update (simple product rule)
updated_belief = round(prior * observation, 2)

# Output
print("Prior:", prior)
print("Observation Likelihood:", observation)
print("Updated Belief:", updated_belief)
```
