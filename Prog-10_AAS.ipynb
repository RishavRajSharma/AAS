```python
import random
import math

# Random gradients
Gx = random.randint(1, 20)
Gy = random.randint(1, 20)

# Edge strength calculation
edge = int(math.sqrt(Gx**2 + Gy**2))

# Output
print("Gx:", Gx, "Gy:", Gy)
print("Edge Strength:", edge)
```
