```python
sensor_metadata = {
    "LiDAR (Velodyne HDL-64E)": {
        "range_meters": 120,
        "dataset_source": "KITTI Vision Benchmark Suite"
    },
    "Stereo Camera": {
        "range_meters": 80,
        "dataset_source": "KITTI Vision Benchmark Suite"
    },
    "Radar": {
        "range_meters": 100,
        "dataset_source": "KITTI Vision Benchmark Suite"
    }
}

# Step 1: Compare sensor ranges
print("Sensor Range Comparison:\n")
for sensor, data in sensor_metadata.items():
    print(f"{sensor} --> Max Range: {data['range_meters']} meters")

# Step 2: Select sensor with maximum range
selected_sensor = max(sensor_metadata, key=lambda x: sensor_metadata[x]["range_meters"])

# Step 3: Output result
print("\nSelected Sensor Based on Maximum Range:")
print(selected_sensor)

# Final Output Requirement
if "LiDAR" in selected_sensor:
    print("\nOutput: LiDAR Selected")
else:
    print("\nOutput: Another Sensor Selected")
```
