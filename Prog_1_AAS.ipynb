```python
import numpy as np
import pandas as pd
from zipfile import ZipFile
import urllib.request
import os


url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00240/UCI%20HAR%20Dataset.zip"
zip_path = "UCI_HAR_Dataset.zip"

if not os.path.exists(zip_path):
    print("Downloading dataset...")
    urllib.request.urlretrieve(url, zip_path)

with ZipFile(zip_path, 'r') as z:
    z.extractall("UCI_HAR_Dataset")


data_path = "UCI_HAR_Dataset/UCI HAR Dataset/test/Inertial Signals/body_acc_z_test.txt"
df_acc_z = pd.read_csv(data_path, delim_whitespace=True, header=None)

print("Loaded motion feature shape:", df_acc_z.shape)

# Take the mean of vertical acceleration as our "altitude proxy"
df_acc_z["altitude_proxy"] = df_acc_z.mean(axis=1)

def classify_robot(altitude_value):
    if altitude_value > 0:
        return "Aerial Autonomous Robot"
    else:
        return "Ground Autonomous Robot"

df_acc_z["robot_type"] = df_acc_z["altitude_proxy"].apply(classify_robot)

sample = df_acc_z[["altitude_proxy", "robot_type"]].head(10)
print("\n=== Sample Output ===")
print(sample.to_string(index=False))

# Show counts
counts = df_acc_z["robot_type"].value_counts()
print("\n=== Counts ===")
print(counts)
```
