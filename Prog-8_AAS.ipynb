```python
import random

f = random.randint(100, 1000)   # focal length
X = random.randint(1, 10)       # horizontal coordinate
Z = random.randint(1, 10)       # depth (avoid 0)

u = (f * X) / Z

print("Focal Length (f):", f)
print("X:", X)
print("Z:", Z)
print("Pixel Coordinate (u):", u, "pixels")
```
