```python
import random

#use given dataset values OR random values
choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    width = 60
    height = 30
else:
    # Random values
    width = random.randint(10, 200)
    height = random.randint(10, 200)

# Compute area
area = width * height

# Output
print("Width:", width, "px")
print("Height:", height, "px")
print("Area:", area, "px²")
```
