```python
import random
import math

print("Go-To-Goal Navigation Initialized")

choice = input("Enter '1' for given coordinates or '2' for random: ")

if choice == '1':
    x1, y1 = 0, 0
    x2, y2 = 6, 8
else:
    # Random coordinates
    x1, y1 = random.randint(0, 10), random.randint(0, 10)
    x2, y2 = random.randint(0, 10), random.randint(0, 10)

# Compute Euclidean distance
distance = round(math.sqrt((x2 - x1)**2 + (y2 - y1)**2), 2)

# Output
print("Start Position:", (x1, y1))
print("Goal Position:", (x2, y2))
print("Distance:", distance, "m")
```
