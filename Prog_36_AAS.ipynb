```python
import random

print("Line Following System Initialized")

choice = input("Enter '1' for given offset or '2' for random: ")

if choice == '1':
    offset = 25
else:
    # Random offset (-50 to +50 pixels)
    offset = random.randint(-50, 50)

# Control logic
if offset > 0:
    action = "Turn Left"
elif offset < 0:
    action = "Turn Right"
else:
    action = "Move Straight"

# Output
print("Offset:", offset, "px")
print("Action:", action)
```
