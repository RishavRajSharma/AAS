```python
import random

# Simulated modules
print("Autonomous System Initialized")

# Option: use fixed success OR random success
choice = input("Enter '1' for successful execution or '2' for random: ")

if choice == '1':
    perception = True
    planning = True
    control = True
else:
    # Random module status
    perception = random.choice([True, False])
    planning = random.choice([True, False])
    control = random.choice([True, False])

# System integration check
if perception and planning and control:
    result = "Autonomous Execution Successful"
else:
    result = "Execution Failed"

# Output
print("Perception:", perception)
print("Planning:", planning)
print("Control:", control)
print("Result:", result)
```
