```python
import random

print("C-Space Model Initialized")

choice = input("Enter '1' for given radius or '2' for random: ")

if choice == '1':
    robot_radius = 0.6
else:
    # Random radius (simulating robot size)
    robot_radius = round(random.uniform(0.2, 1.0), 2)

obstacle_boundary = 1.0

inflated_boundary = obstacle_boundary + robot_radius

# Output
print("Robot Radius:", robot_radius, "m")
print("Original Obstacle Boundary:", obstacle_boundary, "m")
print("Inflated Boundary:", inflated_boundary, "m")
print("Result: Collision-Free Space")
```
