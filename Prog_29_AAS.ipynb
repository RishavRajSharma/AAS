```python
import random

threshold = 20
print("Hybrid Energy System Initialized (Threshold =", threshold, "%)")

choice = input("Enter '1' for given battery level or '2' for random: ")

if choice == '1':
    battery = 18
else:
    # Random battery level
    battery = random.randint(5, 100)

# Decision rule
if battery < threshold:
    mode = "Charging Mode"
else:
    mode = "Operational Mode"

# Output
print("Battery Level:", battery, "%")
print("Mode:", mode)
```
