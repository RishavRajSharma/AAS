```python
import random
import time

print("Patrolling System Initialized")

choice = input("Enter '1' for 5 waypoints or '2' for random: ")

if choice == '1':
    waypoints = [(1,1), (2,2), (3,1), (2,0), (1,0)]
else:
    # Generate random waypoints
    waypoints = [(random.randint(0,5), random.randint(0,5)) for _ in range(5)]

# Patrol loop (1 cycle demo)
print("Starting Patrol...")

for point in waypoints:
    print("Moving to waypoint:", point)
    time.sleep(0.5)  # simulate movement

print("Patrol Completed → Continuing Patrol... (Continuous Patrol)")
```
