```python
import time
import random

rate = 15   # Hz
interval = 1 / rate

print("ROS2 Publisher Simulation Started at", rate, "Hz")

#fixed or random duration
choice = input("Enter '1' for fixed run or '2' for random: ")

if choice == '1':
    duration = 2
else:
    duration = random.randint(1, 5)

start_time = time.time()
count = 0

while time.time() - start_time < duration:
    print("Publishing message", count)
    count += 1
    time.sleep(interval)

# Check performance
actual_rate = count / duration

if actual_rate >= rate:
    result = "Real-Time Performance Achieved"
else:
    result = "Performance Not Achieved"

# Output
print("Messages Sent:", count)
print("Actual Rate:", round(actual_rate, 2), "Hz")
print("Result:", result)
```
