```python
# SESSION 7 – Optimization (JPMorgan)

import sympy as sp

x = sp.symbols('x')

E = x**2 - 6*x + 10

dE_dx = sp.diff(E, x)
print("Derivative dE/dx =", dE_dx)

critical_points = sp.solve(dE_dx, x)
print("Critical Point(s):", critical_points)

min_x = critical_points[0]
min_energy = E.subs(x, min_x)

print("Minimum Energy at x =", min_x)
print("Minimum Energy Value =", min_energy)
```
