```python
import random

print("Wander Bot Initialized")

choice = input("Enter '1' for given angle or '2' for random: ")

if choice == '1':
    angle = 90
else:
    # Random turning angle (0 to 360 degrees)
    angle = random.randint(0, 360)

# Initial direction (in degrees)
direction = 0

# Update direction
new_direction = (direction + angle) % 360

# Output
print("Initial Direction:", direction, "°")
print("Turn Angle:", angle, "°")
print("New Direction:", new_direction, "°")
print("Result: Direction Updated")
```
