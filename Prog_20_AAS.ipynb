```python
import random

threshold = 0.2
print("Navigation System Initialized (Threshold =", threshold, "m)")

choice = input("Enter '1' for given distance or '2' for random: ")

if choice == '1':
    distance = 0.15
else:
    # Random distance (simulating robot navigation)
    distance = round(random.uniform(0.05, 1.0), 2)

# Goal checking
if distance < threshold:
    result = "Goal Reached"
else:
    result = "Goal Not Reached"

# Output
print("Distance to Goal:", distance, "m")
print("Result:", result)
```
