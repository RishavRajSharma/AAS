```python
import random

print("Feedback Control System Initialized")

choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    setpoint = 10
    actual = 7
else:
    # Random values
    setpoint = random.randint(5, 20)
    actual = random.randint(0, 15)

# Compute error
error = setpoint - actual

# Output
print("Setpoint:", setpoint)
print("Actual Value:", actual)
print("Error:", error)
```
