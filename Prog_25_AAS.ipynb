```python
import random

print("Potential Field Model Initialized")

choice = input("Enter '1' for given values or '2' for random: ")

if choice == '1':
    attractive = 12
    repulsive = 5
else:
    # Random forces
    attractive = random.randint(5, 20)
    repulsive = random.randint(1, 10)

net_force = attractive - repulsive

# Output
print("Attractive Force:", attractive)
print("Repulsive Force:", repulsive)
print("Net Force:", net_force)
```
